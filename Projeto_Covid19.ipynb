# Panorama do COVID-19 no Brasil

COVID-19 é uma doença infecciosa causada pelo coronavírus.

Transmitido principalmente por meio de gotículas provenientes de tosses ou espirros de pessoas infectadas, a gravidade dos sintomas varia muito de pessoa para pessoa.

No inicio da pandemia vários comercios, escolas, respartições públicas, Shows e eventos fecharam para o corona virus não se propagar. O uso de mascara era obrigatorio para não propagar virus. Aos poucos o comercio foi reabrindo com o controle de pessoas e cuidados para não proliferação do virus.

Em 2021 foi iniciado a vacinação onde os sintomas do covid-19 foi dominuindo e a mortalidade foi diminuindo e os shows e evento onde circula muitas gentes foi liberando gradualmente, o uso da máscara foi deixando de ser obrigatorio onde a vida foi voltando a normalidade.

Hoje em dia o covid ainda existe, mais com o avanço da vacina a mortalidade em decorrencia do covid-19 está baixando os leitos de uti está diminuindo. As pessoas que está hospitalizado no hospital com sintomas graves da doença 90% dessas pessoas não completaram o esquema vacinal.

*  `iso_code` - Código do país em 3 letras.

*  `date` - Data dos dados obtidos
*  `total_cases` - Total de casos confirmados
*  `new_cases` - Novos casos confirmados
*  `new_cases_smoothed` - Novos casos cofirmaos nivelados por 7 dias
*  `total_deaths` - Total de Mortes confirmadas em decorrencia da covid 19
*  `new_deaths` - Novas Mortes confimrmadas em decorrencia da covid 19
*  `new_deaths_smoothed` - Novas Mortes cofirmaos nivelados por 7 dias
*  `total_cases_per_million` - Total de casos confirmados por 1.000.000 de pessoas.
*  `new_cases_per_million` - Novos casos confirmados por 1.000.000 de pessoas.
*  `new_cases_smoothed_per_million` - Novos casos confirmados nivelados por 1.000.000 de pessoas.
*  `total_deaths_per_million` - Total de Mortes confirmadas em decorrencia da covid 19 por 1.000.000 de pessoas.
* `new_deaths_per_million` - Novas Mortes confirmadas em decorrencia da covid 19 por 1.000.000 de pessoas.
*  `new_deaths_smoothed_per_million` - Novas Mortes confirmadas em decorrencia da  covid 19 niveladas em 7 dias por 1.000.0000 de pessoas
*  `reproduction_rate` - Estimativa em tempo real da proporção de reprodução do Covid-19
*  `icu_patients` - Número de paciente com Covid-19 em Unidades de Terapia Intensiva (UTI) em um determinado dia.
* `icu_patients_per_million` - Número de paciente com Covid-19 em Unidades de Terapia Intensiva (UTI) em um determinado dia por 1.000.000 de pessoas.
* `hosp_patients` - Número de pacientes com Covid-19 hospitalizados em determinado dia.
* `hosp_patients_per_million` - Número de pacientes com Covid-19 por 1.000.000 de pessoas.
* `weekly_icu_admissions` - Número de pacientes com Covid-19 admitidos para a UTI em determinada semana.
* `weekly_icu_admissions_per_million` - Número de pacientes com Covid-19 admitidos para a UTI em determinada semana por 1.000.000 de pessoas.
* `weekly_hosp_admissions` - Número de pacientes com Covid-19 admitidos em hospitais recentemente, em determinada semana.
* `weekly_hosp_admissions_per_million` - Número de pacientes com Covid-19 admitidos em hospitais recentemente determinada semana por 1000000 de pessoas.
* `total_tests` - Número total de testes para Covid-19.
* `new_tests` - Novos testes para Covid-19.
* `total_tests_per_thousand` - Número total de testes para Covid-19 por 1.000 pessoas.
* `new_tests_per_thousand` - Novos testes para Covid-19 por 1.000 pessoas.
* `new_tests_smoothed` - Novos testes para Covid-19 nivelados por 7 dias.
* `new_tests_smoothed_per_thousand` - Novos testes para Covid-19 nivelados por 7 dias por 1.000 pessoas.
* `positive_rate` - Proporção dos testes para Covid-19 que são positivos dentro de uma média de 7 dias.
* `tests_per_case` - Testes conduzidos por novos casos confirmados de Covid-19 dentro de uma média de 7 dias.
* `tests_units` - Quantidade de testes utilizado por localização
* `total_vaccinations` - Número total de doses de vacinas de Covid-19.
* `people_vaccinated` - Número total de pessoas que receberam pelo menos uma dose de vacina de Covid-19.
* `people_fully_vaccinated` - Número total de pessoas que receberam todas as doses de vacina por determinado protocolo de vacinação.
* `total_boosters` - Número total de doses de reforço da vacina.
* `new_vaccinations` - Novas vacinações.
* `new_vaccinations_smoothed` - Novas vacinações niveladas por 7 dias
* `total_vaccinations_per_hundred` - Número total de vacinações a cada 100 pessoas.
* `people_vaccinated_per_hundred` - Número total de pessoas que receberam pelo menos uma dose de vacina a cada 100 pessoas.
* `people_fully_vaccinated_per_hundred` - Número total de pessoas que receberam todas as doses de vacina por determinado protocolo de vacinação a cada 100 pessoas.
* `total_boosters_per_hundred` - Número total de doses de reforço administradas a cada 100 pessoas.
* `new_vaccinations_smoothed_per_million` - Novas doses de vacina administradas niveladas por 7 dias por 1.000.000 pessoas.
* `new_people_vaccinated_smoothed` - Número diário de pessoas recebendo a primeira dose da vacina nivelado por 7 dias.
* `new_people_vaccinated_smoothed_per_hundred` - Número diário de pessoas recebendo a primeira dose da vacina nivelado por 7 dias por 100 pessoas.
* `stringency_index` - Índice de Rigidez de Resposta do Governo: medida composta com base em 9 indicadores de resposta, incluindo fechamento de escolas, fechamento de locais de trabalho e proibição de viagens, reescalonado para um valor de 0 a 100 (100 = resposta mais rigorosa).
* `population` - População.
* `population_density` - Número de pessoas dividida pela área, medida em quilômetros quadrados.
* `median_age` - Idade média da população.
* `aged_65_older` - Corte da população com idade de 65 anos ou mais velha.
* `aged_70_older` - Corte da população com idade de 70 anos ou mais velha.
* `gdp_per_capita` - Produto Interno Bruto em paridade com poder de compra (em dólares internacionais).
* `extreme_poverty` - Porção da população vivendo em extrema pobreza, data mais recente disponível: desde 2010.
* `cardiovasc_death_rate` - Proporção de morte por doenças cardiovasculares em 2017.
* `diabetes_prevalence` - Prevalência de diabetes (% da população de 20 a 79 anos) em 2017
* `female_smokers` - Proporção de mulheres que fumam, ano mais recente disponível
* `male_smokers` - Proporção de homens que fumam, ano mais recente disponível
* `handwashing_facilities` - Percentagem da população com instalações básicas de lavagem das mãos nas instalações, ano mais recente disponível
* `hospital_beds_per_thousand` - Camas hospitalares por 1.000 pessoas, ano mais recente disponível desde 2010 
* `life_expectancy` - Expectativa de vida ao nascer em 2019
* `human_development_index` - Um índice composto que mede o desempenho médio em três dimensões básicas do desenvolvimento humano – uma vida longa e saudável, conhecimento e um padrão de vida decente. Valores para 2019.
* `excess_mortality_cumulative_absolute` - Diferença percentual entre o número relatado de mortes semanais ou mensais em 2020-2021 e o número projetado de mortes para o mesmo período com base em anos anteriores.
* `excess_mortality_cumulative` - Diferença percentual entre o número acumulado de mortes desde 1 de janeiro de 2020 e as mortes cumulativas projetadas para o mesmo período com base em anos anteriores
* `excess_mortality` -  Diferença cumulativa entre o número relatado de óbitos desde 1º de janeiro de 2020 e o número projetado de óbitos para o mesmo período com base em anos anteriores. 
* `excess_mortality_cumulative_per_million` - Diferença cumulativa entre o número relatado de mortes desde 1º de janeiro de 2020 e o número projetado de mortes para o mesmo período com base em anos anteriores, por milhão de pessoas.



### Importando os dados do COVID-19



```python
# importar as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style()
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

# importar o dataset COVID-19 para um DataaFrame
df = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv')
```

## Análise Exploratória dos Dados

Iremos iniciar a análise exploratória onde veremos até a data 25/06/2022. Quais países com mais casos confirmados de covid, mortes confirmadas, vacinação. Como está os casos de covid no mundo.

```python
# Visualizar as 5 primeiras entradas
df.head()
```

```python
# Preenchendo NaN de new_deaths com 0
df.new_deaths.fillna(0, inplace=True)
df.new_cases.fillna(0, inplace=True)
```

```python
#Extrair e imprimir os nomes das colunas
df.columns
```

```python
#Qual o tamanho do seu DataFrame
print('Dimensões do DataSet', '\nVariáveis:',
df.shape[1], '\nEntradas:', df.shape[0])
```

```python
#Quais os tipos das variáveis
df.dtypes
```

```python
#porcentagem de valores ausentes das colunas
round((df.isnull().sum()/ df.shape[0]).sort_values(ascending=False), 2)
```

```python
#Transforme a coluna `date` no formato `datetime`
df.date = pd.to_datetime(df.date)
```

```python
# Ver para a data mais atual.
df.date.max()
```

```python
# quais os Continentes que têm mais casos confirmados
df.loc[df.date == '2022-06-25'].sort_values(by = 'total_cases', ascending=False) [2:7]
```

Podemos ver que o vírus está em todo o mundo e em todos o continentes não diferenciando de classes sociais (países ricos, pobres e emergente), onde os países com mais casos confirmados está na Europa (203.598.634 Casos confirmados), Asia(156.234.953 casos confirmados), União Europeia(147.229.619 casos confirmados). Países onde o PIB é maior e onde a população tem um poder aquisitivo mais alto.



```python
# Paises com mais casos confirmados
df.loc[df.date == '2022-06-25'].sort_values(by = 'total_cases', ascending=False) [8:15]
```

```python
# Gráfico com mais casos Confirmados
top_total_casos= df.loc[df.date=='2022-06-09', ['location', 'total_cases']].sort_values(by='total_cases', ascending=False)[8:15]
fig, ax =plt.subplots(figsize=(15,10))
ax.bar('location', 'total_cases', data=top_total_casos)
plt.tight_layout()
```

Podemos ver que os países com mais casos confirmados são Estados Unidos, Índia e Brasil. São países com grandes populações. Logo em seguida vem países do continente Europeu com França, Alemanha e Reino Unido onde são países com PIB elevados, onde só verificamos que esse vírus não tem classes sóciais.

```python
# Paises com mais mortes
df.loc[df.date == '2022-06-25'].sort_values(by= 'total_deaths', ascending=False) [9:16]
```

```python
top_total_deaths= df.loc[df.date=='2022-06-25', ['location', 'total_deaths']].sort_values(by='total_deaths', ascending=False)[9:14]
fig, ax =plt.subplots(figsize=(15,10))
ax.bar('location', 'total_deaths', data=top_total_deaths)
plt.tight_layout()
```

Podemos ver que os Estados Unidos é o país com mais casos confirmados e mortes sendo o pais que mais sofreu com a covid-19 em seguida vem o Brasil que mais mortes que a Índia. 
Então podemos ver que a quantidade de casos não está relacionado a mortes. A Índia tem mais casos confirmados que o Brasil só que o Brasil tem mais mortes. Podemos ver que países trata a questão de lockdown sistema hospitalar, saúde de seus habitantes tem menos casos de mortes. 
Podemos ver que Europa tem países com muitos casos confirmados, mais as mortes não está entre os grandes. Rússia e Mexico aparece no top 5 de países com mais mortes, mais não aparece em casos confirmados então isso tem relação com alguns fatores relacionados acima.

```python
#Paises Com mais  Pessoas Vacinadas com pelo menos 1 dose Dia 25/06/2022
df.loc[df.date == '2022-06-25'].sort_values(by = 'people_vaccinated', ascending=False) [4:15]
```

```python
#Paises Com mais  Pessoas Vacinadas com pelo menos 1 dose Dia 01/06/2022
df.loc[df.date == '2022-06-01'].sort_values(by = 'people_vaccinated', ascending=False) [4:15]
```

Como podemos ver nos dataset acima um dia 25/06/2022 e outro  dia 01/06/2022
onde tem variantes de pessoas vacinadas no dia 01/06/2022 o top 3 é china, Índia e Estados Unidos e no dia 25/06/2022 é India, Brasil e Paquistão onde é bastante variável os países com pelo menos 1 dose aplicada. Não temos como tirar grande conclusões pois é bastante variável.

```python
df_wold = df.loc[df.location == 'World'].copy()
df_wold.head()
```

```python
x = df_wold['date']
z = df_wold['total_cases']
z2 = df_wold['total_deaths']
z3 = df_wold['people_vaccinated']

fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1, figsize=(10,7))

# plotar total De casos
ax1.set_title("Gráfico de casos no mundo")
ax1.set_ylabel("Casos Confirmados")
ax1.set_xlabel("Data")
ax1.plot(x, z)

# plotar total de mortes
ax2.set_title("Gráfico de Mortes")
ax2.set_ylabel("Casos Confirmados")
ax2.set_xlabel("Data")
ax2.plot(x, z2)

# plotar pessoas vacinadas com pelo menos 1 dose
ax3.set_title("Gráfico de Pessoas Vacinadas")
ax3.set_ylabel("Casos Confirmados")
ax3.set_xlabel("Dias")
ax3.plot(x, z3)

plt.tight_layout()
```

### Análise Exploratória para o Brasil
Agora vamos ver um pouco os dados de covid-19 no Brasil.

```python
# Criar uma cópia do DataFrame apenas com o Brasil
df_brasil = df.loc[df.location == "Brazil"].copy()

df_brasil.head()
```

```python
# Preenchendo NaN de new_deaths com 0
df_brasil.new_deaths.fillna(0, inplace=True)
df_brasil.new_cases.fillna(0, inplace=True)
```

```python
df_brasil.date.max()
```

Podemos ver que esse dataset é do dia 26/02/2020 onde teve o 1 caso de covid confirmado no país ate o dia 25/06/2022

```python
# primeira morte
df_brasil.loc[df_brasil['new_deaths'] != 0.0].head(1)
```

```python
# Identificar quantos dias demorou para registrarmos a primeira morte, considerando o primeiro caso de COVID-19 registrado.
# calculando tempo desde o primeiro caso até a primeira morte
from datetime import date
delta = df_brasil.date.loc[24572] - df_brasil.date.loc[24552]
print('Se passaram {} dias entre o primeiro caso e a primeira morte no Brasil.'.format(delta.days))
```

```python
# Plotar gráficos de linha em escalas linear e logarítmica para os casos totais e mortes totais.
x = df_brasil['date']
z = df_brasil['total_cases']
z2 = df_brasil['total_deaths']

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(15, 7))
ax1.set_title('Casos no Brasil')
ax1.set_xlabel('Data')
ax1.set_ylabel('Casos')
ax1.plot(x, z)

ax2.set_title('Casos no Brasil (Escala logarítmica)')
ax2.set_xlabel('Data')
ax2.set_ylabel('Casos')
ax2.set_yscale('log')
ax2.plot(x, z)

ax3.set_title('Mortes no Brasil')
ax3.set_xlabel('Data')
ax3.set_ylabel('Mortes')
ax3.plot(x, z2)

ax4.set_title('Mortes no Brasil (Escala logarítmica)')
ax4.set_xlabel('Dias')
ax4.set_ylabel('Mortes')
ax4.set_yscale('log')
ax4.plot(x, z2)
plt.tight_layout()
```

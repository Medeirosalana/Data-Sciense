# Obtenção dos Dados

Vamos iniciar nossa obtenção de dados com as importações dos
arquivos necessarios para iniciar a analise dos dados.

```python
# Importação dos arquivos
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Importação arquivo csv 
df=pd.read_csv("http://data.insideairbnb.com/united-states/ca/san-francisco/2021-12-04/visualisations/listings.csv")
```

# Análise dos dados

Vamos iniciar a analise dos dados obtidos do arquivo.

**Dicionário dos dados**


*   `id` - Número da chave para indentificar o imóvel
*   `name`- Nome da propriedade anunciada
*   `host_id` - Número da chave do proprietário da propriedade
*   `host_name` - Nome do proprietário
*   `neighbourhood_group` - Essa columa não possui nenhum valor válido
*   `neighbourhood` - Bairro da propriedade
*   `latitude` - Coordenada da latitude da propriedade
*   `longitude` - Coordenada da longitude da propriedade
*   `room_type` - Tipo de quarto ofertado
*   `price` - Preço para alugar o imóvel
*   `minimum_nights` - Quantidade minina de noites para reservar
*   `number_of_reviews` - Número de visitas que o imovel possui
*   `last_review` - Data da ultima visita
*   `reviews_per_month` - Número de visitas por més 
*   `calculated_host_listings_count` - Quantidade de imoveis do mesmo proprietario
*   `availability_365` - Número de dias de disponilidade em 365 dias
*   `number_of_reviews_ltm` - Número de avaliação do imóvel em 12 meses 
*   `license` - Autorização ou número de registro do imóvel

Agora vamos ver as 5 primeiras entradas do arquivo

# Quantidade de Variaveis e entradas e os tipos da variáveis

Vamos prosseguir e identificar a quantidade de entradas que nosso conjunto de dados possui e ver os tipos de cada coluna.

```python
# Identificar o volume de dados
print("Entradas:\t{}".format(df.shape[0]))
print('Variáveis:\t{}\n'.format(df.shape[1]))

# Verificar as 5 primeiras entradas do dataset
display(df.dtypes)
```

# Porcentagem de valores ausentes

É importante verificar os valores ausentes nas linhas para analizar se os valores nulos são significativos em relação as entradas. Visto que esses valores ausentes fazem difirença no final da analise. 

*   Verificamos que `neighbourhood_group` possui 100% dos dados nulos
*   Verificamos que `license` possui quase 50% dos dados nulos
*   Verificamos que `reviews_per_month` e `last_review` possui 20% dos dados nulos






```python
# ordenar em ordem decrescente as variáveis por seus valores ausentes
(df.isnull().sum()/ df.shape[0]).sort_values(ascending=False)
```

# Distribuição das variaveis

Vamos colocar no histograma os dados do arquivos

```python
# plotar o histograma das variáveis numéricas
df.hist(bins=15 , figsize=(15,10));
```

# Media de preços

Pedemos ver que a media de preços em São Francis é de 228,63 dólares

```python
# Media de preços dos alugueis
df.price.mean()
```


# Correlação entre as variaveis

Vejamos agora a correlação entre as variaveis, ou seja, vamos ver entre as variaveis se tem semelança ou relação.

```python
# criar uma matriz de correlação
corr = df[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month',
    'calculated_host_listings_count', 'availability_365']].corr()
# mostrar a matriz de correlação
display(corr)
```

```python
# plotar um heatmap a partir das correlações
sns.heatmap(corr, cmap='RdBu', fmt='.2f', square=True, linecolor='white', annot=True);
```

# Tipo de imovel mais alugado

No Airbnb podemos alugar diversos tipos de imovel como um apartamento/casas inteiras somente um quarto ou dividir com pessoas.


```python
# mostrar a quantidade de cada tipo de imóvel disponível
df.room_type.value_counts()
```

```python
# mostrar a porcentagem de cada tipo de imóvel disponível
df.room_type.value_counts() / df.shape[0]
```

# Localidade mais cara

Vejamos os bairros mais caro de São Francisco 

```python
# ver preços por bairros, na média
df.groupby(['neighbourhood']).price.mean().sort_values(ascending=False)[:10]
```

Podemos ver que os bairros mais caros de São Francisco é Financial District, chinatown e Russian Hill

Agora com a latitude e longitude podemos ver onde fica localizado cada imóvel.

```python
# plotar os imóveis pela latitude-longitude
df.plot(kind="scatter", x='longitude', y='latitude', alpha=0.4, c=df['price'], s=8,
              cmap=plt.get_cmap('jet'), figsize=(12,8));
```

# Media minima de noites
